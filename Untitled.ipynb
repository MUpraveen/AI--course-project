```python
import numpy as np
import matplotlib.pyplot as plt
from collections import deque
import heapq
import time

class ConvexHullPathfinding:
    def __init__(self, grid_width=800, grid_height=600, num_obstacles=5):
        self.grid_width = grid_width
        self.grid_height = grid_height
        self.num_obstacles = num_obstacles
        self.start = (50, 300)
        self.goal = (750, 450)
        self.obstacles = []
        self.vertices = []
        self.valid_edges = []
        
    def graham_scan(self, points):
        """Graham Scan algorithm to find convex hull"""
        if len(points) < 3:
            return points
        
        # Find the point with lowest y-coordinate (and leftmost if tie)
        p0 = min(points, key=lambda p: (p[1], p[0]))
        
        # Sort points by polar angle with respect to p0
        def angle(p):
            return np.arctan2(p[1] - p0[1], p[0] - p0[0])
        
        sorted_points = sorted([p for p in points if p != p0], key=lambda p: (angle(p), self.dist(p0, p)))
        
        # Build convex hull
        hull = [p0, sorted_points[0]]
        
        def ccw(a, b, c):
            return (b[0] - a[0]) * (c[1] - a[1]) - (b[1] - a[1]) * (c[0] - a[0])
        
        for p in sorted_points[1:]:
            while len(hull) > 1 and ccw(hull[-2], hull[-1], p) <= 0:
                hull.pop()
            hull.append(p)
        
        return hull
    
    def generate_obstacles(self):
        """Generate random convex polygonal obstacles using Graham Scan"""
        self.obstacles = []
        margin = 150
        
        for _ in range(self.num_obstacles):
            num_points = np.random.randint(5, 10)
            center_x = margin + np.random.rand() * (self.grid_width - 2 * margin)
            center_y = margin + np.random.rand() * (self.grid_height - 2 * margin)
            radius = 30 + np.random.rand() * 40
            
            # Generate random points around center
            points = []
            for _ in range(num_points):
                angle = np.random.rand() * 2 * np.pi
                r = radius * (0.5 + np.random.rand() * 0.5)
                x = center_x + r * np.cos(angle)
                y = center_y + r * np.sin(angle)
                points.append((x, y))
            
            # Apply Graham Scan to get convex hull
            hull = self.graham_scan(points)
            self.obstacles.append(hull)
    
    def segments_intersect(self, p1, p2, p3, p4):
        """Check if line segment p1-p2 intersects with p3-p4"""
        def ccw(A, B, C):
            return (C[1] - A[1]) * (B[0] - A[0]) > (B[1] - A[1]) * (C[0] - A[0])
        
        # Check if segments are identical or share endpoints
        if (p1 == p3 and p2 == p4) or (p1 == p4 and p2 == p3):
            return False
        if p1 == p3 or p1 == p4 or p2 == p3 or p2 == p4:
            return False
            
        return ccw(p1, p3, p4) != ccw(p2, p3, p4) and ccw(p1, p2, p3) != ccw(p1, p2, p4)
    
    def point_in_polygon(self, point, polygon):
        """Check if a point is inside a polygon using ray casting"""
        x, y = point
        n = len(polygon)
        inside = False
        
        p1x, p1y = polygon[0]
        for i in range(1, n + 1):
            p2x, p2y = polygon[i % n]
            if y > min(p1y, p2y):
                if y <= max(p1y, p2y):
                    if x <= max(p1x, p2x):
                        if p1y != p2y:
                            xinters = (y - p1y) * (p2x - p1x) / (p2y - p1y) + p1x
                        if p1x == p2x or x <= xinters:
                            inside = not inside
            p1x, p1y = p2x, p2y
        
        return inside
    
    def edge_intersects_obstacle(self, v1, v2, obstacle):
        """Check if edge v1-v2 intersects with any edge of the obstacle or passes through it"""
        # Check if edge intersects with any obstacle edge
        for i in range(len(obstacle)):
            o1 = obstacle[i]
            o2 = obstacle[(i + 1) % len(obstacle)]
            if self.segments_intersect(v1, v2, o1, o2):
                return True
        
        # Check if midpoint is inside obstacle
        midpoint = ((v1[0] + v2[0]) / 2, (v1[1] + v2[1]) / 2)
        if self.point_in_polygon(midpoint, obstacle):
            return True
            
        return False
    
    def is_valid_edge(self, v1, v2):
        """Check if edge between v1 and v2 is valid (doesn't intersect obstacles)"""
        if v1 == v2:
            return False
            
        # Check if v1 and v2 are on the same obstacle (then edge is allowed)
        for obstacle in self.obstacles:
            if v1 in obstacle and v2 in obstacle:
                # Check if they are consecutive vertices
                idx1 = obstacle.index(v1)
                idx2 = obstacle.index(v2)
                if abs(idx1 - idx2) == 1 or abs(idx1 - idx2) == len(obstacle) - 1:
                    return True
        
        # Check if edge intersects with any obstacle
        for obstacle in self.obstacles:
            if self.edge_intersects_obstacle(v1, v2, obstacle):
                return False
        
        return True
    
    def build_graph(self):
        """Build graph from convex hull vertices with valid edges only"""
        self.vertices = [self.start, self.goal]
        
        # Add all obstacle vertices
        for obstacle in self.obstacles:
            self.vertices.extend(obstacle)
        
        # Build valid edges (only along obstacle perimeters and non-intersecting connections)
        self.valid_edges = []
        
        # Add edges along obstacle perimeters
        for obstacle in self.obstacles:
            for i in range(len(obstacle)):
                v1 = obstacle[i]
                v2 = obstacle[(i + 1) % len(obstacle)]
                self.valid_edges.append((v1, v2))
        
        # Add valid connections between different obstacles and start/goal
        for i, v1 in enumerate(self.vertices):
            for j, v2 in enumerate(self.vertices):
                if i >= j:
                    continue
                
                # Skip if already added as obstacle edge
                if (v1, v2) in self.valid_edges or (v2, v1) in self.valid_edges:
                    continue
                
                # Check if this edge is valid
                if self.is_valid_edge(v1, v2):
                    self.valid_edges.append((v1, v2))
    
    def get_neighbors(self, vertex):
        """Get neighboring vertices connected by valid edges"""
        neighbors = []
        for v1, v2 in self.valid_edges:
            if v1 == vertex:
                neighbors.append(v2)
            elif v2 == vertex:
                neighbors.append(v1)
        return neighbors
    
    def dist(self, a, b):
        """Euclidean distance"""
        return np.sqrt((a[0] - b[0])**2 + (a[1] - b[1])**2)
    
    def manhattan(self, a, b):
        """Manhattan distance"""
        return abs(a[0] - b[0]) + abs(a[1] - b[1])
    
    def diagonal(self, a, b):
        """Diagonal distance (Chebyshev)"""
        dx = abs(a[0] - b[0])
        dy = abs(a[1] - b[1])
        return max(dx, dy) + 0.414 * min(dx, dy)
    
    def a_star(self, heuristic_func, heuristic_name):
        """A* search with given heuristic"""
        start_time = time.time()
        
        open_set = []
        heapq.heappush(open_set, (0, self.start))
        came_from = {}
        g_score = {v: float('inf') for v in self.vertices}
        g_score[self.start] = 0
        f_score = {v: float('inf') for v in self.vertices}
        f_score[self.start] = heuristic_func(self.start, self.goal)
        
        nodes_explored = 0
        
        while open_set:
            current = heapq.heappop(open_set)[1]
            nodes_explored += 1
            
            if current == self.goal:
                # Reconstruct path
                path = [current]
                while current in came_from:
                    current = came_from[current]
                    path.append(current)
                path.reverse()
                
                elapsed_time = time.time() - start_time
                path_length = sum(self.dist(path[i], path[i+1]) for i in range(len(path)-1))
                
                return {
                    'path': path,
                    'nodes_explored': nodes_explored,
                    'time': elapsed_time,
                    'path_length': path_length
                }
            
            for neighbor in self.get_neighbors(current):
                tentative_g = g_score[current] + self.dist(current, neighbor)
                
                if tentative_g < g_score[neighbor]:
                    came_from[neighbor] = current
                    g_score[neighbor] = tentative_g
                    f_score[neighbor] = tentative_g + heuristic_func(neighbor, self.goal)
                    heapq.heappush(open_set, (f_score[neighbor], neighbor))
        
        return None
    
    def run_comparison(self):
        """Run pathfinding with all three heuristics"""
        self.generate_obstacles()
        self.build_graph()
        
        heuristics = {
            'Manhattan': self.manhattan,
            'Euclidean': self.dist,
            'Diagonal': self.diagonal
        }
        
        results = {}
        for name, func in heuristics.items():
            result = self.a_star(func, name)
            if result:
                results[name] = result
                print(f"\n{name} Distance Heuristic:")
                print(f"  Nodes explored: {result['nodes_explored']}")
                print(f"  Path length: {result['path_length']:.2f}")
                print(f"  Time: {result['time']*1000:.4f} ms")
        
        return results
    
    def visualize(self, results):
        """Visualize the grid, obstacles, and paths"""
        fig, axes = plt.subplots(2, 2, figsize=(15, 12))
        
        # Plot all obstacles in each subplot
        for idx, (name, result) in enumerate(results.items()):
            row = idx // 2
            col = idx % 2
            ax = axes[row, col]
            
            # Draw obstacles (filled)
            for obstacle in self.obstacles:
                poly = plt.Polygon(obstacle, alpha=0.4, facecolor='lightgray', edgecolor='black', linewidth=2)
                ax.add_patch(poly)
            
            # Draw all valid edges (light blue)
            for v1, v2 in self.valid_edges:
                ax.plot([v1[0], v2[0]], [v1[1], v2[1]], 'b-', alpha=0.15, linewidth=0.5)
            
            # Draw vertices
            all_verts = np.array(self.vertices)
            ax.scatter(all_verts[:, 0], all_verts[:, 1], c='blue', s=20, alpha=0.5, zorder=3)
            
            # Draw path (thick red line)
            path = result['path']
            path_x = [p[0] for p in path]
            path_y = [p[1] for p in path]
            ax.plot(path_x, path_y, 'r-', linewidth=3, label='Path', zorder=4)
            
            # Draw start and goal
            ax.scatter(*self.start, c='green', s=200, marker='*', label='Start', zorder=5, edgecolors='black', linewidths=2)
            ax.scatter(*self.goal, c='red', s=200, marker='*', label='Goal', zorder=5, edgecolors='black', linewidths=2)
            
            ax.set_xlim(0, self.grid_width)
            ax.set_ylim(0, self.grid_height)
            ax.set_aspect('equal')
            ax.grid(True, alpha=0.3)
            ax.legend()
            ax.set_title(f'{name} Heuristic\nNodes: {result["nodes_explored"]}, Length: {result["path_length"]:.2f}')
        
        # Performance comparison in 4th subplot
        ax = axes[1, 1]
        if len(results) == 3:
            axes[1, 1].remove()
            ax = fig.add_subplot(2, 2, 4)
            
            names = list(results.keys())
            nodes = [results[n]['nodes_explored'] for n in names]
            times = [results[n]['time']*1000 for n in names]
            lengths = [results[n]['path_length'] for n in names]
            
            x = np.arange(len(names))
            width = 0.25
            
            ax2 = ax.twinx()
            ax3 = ax.twinx()
            ax3.spines['right'].set_position(('outward', 60))
            
            p1 = ax.bar(x - width, nodes, width, label='Nodes Explored', color='skyblue')
            p2 = ax2.bar(x, times, width, label='Time (ms)', color='lightcoral')
            p3 = ax3.bar(x + width, lengths, width, label='Path Length', color='lightgreen')
            
            ax.set_xlabel('Heuristic')
            ax.set_ylabel('Nodes Explored', color='skyblue')
            ax2.set_ylabel('Time (ms)', color='lightcoral')
            ax3.set_ylabel('Path Length', color='lightgreen')
            
            ax.set_xticks(x)
            ax.set_xticklabels(names)
            ax.tick_params(axis='y', labelcolor='skyblue')
            ax2.tick_params(axis='y', labelcolor='lightcoral')
            ax3.tick_params(axis='y', labelcolor='lightgreen')
            
            ax.set_title('Performance Comparison')
            ax.legend(loc='upper left')
            ax2.legend(loc='upper center')
            ax3.legend(loc='upper right')
        
        plt.tight_layout()
        plt.savefig('convex_hull_pathfinding.png', dpi=150, bbox_inches='tight')
        plt.show()

# Run the pathfinding comparison
if __name__ == "__main__":
    pathfinder = ConvexHullPathfinding(grid_width=800, grid_height=600, num_obstacles=9)
    results = pathfinder.run_comparison()
    pathfinder.visualize(results)
```
